```python
class Node:
    def __init__(self, element, next = None):
        self._element = element
        self._next = next

class SLList:

    def __init__(self):
        self._head = None
        self._tail = None
        self._size = 0
    
    def __len__(self):
        return self._size

    def add_first(self, e):
        newest = Node(e, self._head)
        self._head = newest
        self._size += 1
        if self._tail == None:
            self._tail = newest
    
    def add_last(self, e):
        newest = Node(e)
        if self._tail == None:
            self._head = newest
        else: 
            self._tail._next = newest
        self._tail = newest
        self._size += 1
            
    
    def remove_first(self):
        if self._head == None:
            raise Exception("Empty list")
        self._head = self._head._next
        self._size -= 1
        if self._head == None:
            self._tail = None
    
    def find(self, e):
        p = self._head
        while p != None:
            if p._element == e:
                return True
            p = p._next
        return False
    
    def remove(self, e):
        p = self._head
        if p._element == e:
            self._remove_first()
        else:
            while p != None and p._element != e:
                p1 = p
                p = p._next
            if p != None:
                p1._next = p._next
                if p == self._tail:
                    self._tail = p1
                self._size -= 1
        
    def sort(self):
        if self._head == None:
            return
        p = self._head
        while p != self._tail:
            q = p._next
            while q != None:
                if p._element > q._element:
                    tmp = p._element
                    p._element = q._element
                    q._element = tmp
                q = q._next
            p = p._next
    
    def __str__(self):
        s = "["
        p = self._head
        while p != None:
            s = s + str(p._element)
            if p != self._tail:
                s = s + ", "
            
            p = p._next
        s = s + "]"
        return s
class Student:

    def __init__(self, name, age, gpa):
        self._name = name
        self._age = age
        self._gpa = gpa
    
    def __eq__(self, other):
        return self._name == other._name and self._age == other._age and self._gpa == other._gpa

    def __str__(self):
        return "Student[" + self._name + ", " + str(self._age) + ", " + str(self._gpa) + "]"
class Student_List(SLList):
    def findByName(self, name):
        p = self._head
        while p != None:
            if p._element._name == name:
                return p._element
            p = p._next
        return None
    def listGoodStudents(self):
        result = Student_List()
        p = self._head
        while p != None:
            if p._element._gpa >= 8:
                result.add_last(p._element)
            p = p._next
        return result
    def sortByGPA(self):
        p = self._head
        while p != None:
            q = p._next
            while q != None:
                if p._element._gpa > q._element._gpa:
                    p._element, q._element = q._element, p._element
                q = q._next
            p = p._next
    def addByGPA(self,s):
        if self._head is None or s._gpa >= self._head._element._gpa:
            self.add_first(s)
        else:
            p = self._head
            p1 = None
            while p != None and p._element._gpa > s._gpa:
                p1 = p
                p = p._next
            q = Node(s,p)
            p1._next = q
            self._size += 1
    def append_list(self, other):
        if other._size == 0:
            return
        elif self._size == 0:
            self._head = other._head
            self._tail = other._tail
            self._size = other._size
        else:
            self._tail._next = other._head
            self._tail = other._tail
            self._size += other._size
        other._head = None
        other._tail = None
        other._size = 0
    def merge_list_by_GPA(self, other):
        new_list = Student_List()

        curr1 = self._head
        curr2 = other._head
        while curr1 and curr2:
            if curr1._element._gpa >= curr2._element._gpa:
                new_list.add_last(curr1._element)
                curr1 = curr1._next
            else:
                new_list.add_last(curr2._element)
                curr2 = curr2._next
        while curr1:
            new_list.add_last(curr1._element)
            curr1 = curr1._next
        while curr2:
            new_list.add_last(curr2._element)
            curr2 = curr2._next
        return new_list
    

l2 = Student_List()
l2.add_last(Student('Anh', 19, 7.5))
l2.add_last(Student('Binh', 21, 8.3))
l3 = Student_List()
l3.add_last(Student('Teo', 22, 9.1))
l3.add_last(Student('Ti', 20, 9.0))

print(l2.merge_list_by_GPA(l3))
```

```python
class PolyNode:
    def __init__(self, coef, exp, next=None):
        self._coef = coef
        self._exp = exp
        self._next = next

class Poly:
    def __init__(self):
        self._head = None

    def __str__(self):
        """Phương thức chuyển đa thức thành xâu ký tự"""
        if self._head is None:
            return "0"
        else:
            result = ""
            current_node = self._head
            while current_node is not None:
                if current_node._coef > 0 and result != "":
                    result += "+"
                if current_node._coef != 1:
                    result += str(current_node._coef)
                if current_node._exp > 0:
                    result += "x"
                if current_node._exp > 1:
                    result += "^" + str(current_node._exp)
                current_node = current_node._next
            return result

    def evaluate(self, x):
        """Phương thức tính giá trị đa thức"""
        result = 0
        current_node = self._head
        while current_node is not None:
            result += current_node._coef * (x ** current_node._exp)
            current_node = current_node._next
        return result

    def __add__(self, other):
        """Phương thức cộng hai đa thức"""
        new_poly = Poly()
        current_node1 = self._head
        current_node2 = other._head
        while current_node1 is not None or current_node2 is not None:
            if current_node1 is None:
                new_poly.insert_node(current_node2._coef, current_node2._exp)
                current_node2 = current_node2._next
            elif current_node2 is None:
                new_poly.insert_node(current_node1._coef, current_node1._exp)
                current_node1 = current_node1._next
            else:
                if current_node1._exp > current_node2._exp:
                    new_poly.insert_node(current_node1._coef, current_node1._exp)
                    current_node1 = current_node1._next
                elif current_node2._exp > current_node1._exp:
                    new_poly.insert_node(current_node2._coef, current_node2._exp)
                    current_node2 = current_node2._next
                else:
                    new_coef = current_node1._coef + current_node2._coef
                    if new_coef != 0:
                        new_poly.insert_node(new_coef, current_node1._exp)
                    current_node1 = current_node1._next
                    current_node2 = current_node2._next
        return new_poly


    def insert_node(self, coef, exp):
        """Phương thức chèn một node vào đúng vị trí giảm dần của số mũ"""
        new_node = PolyNode(coef, exp)
        if self._head is None or exp > self._head._exp:
            new_node._next = self._head
            self._head = new_node
        else:
            current_node = self._head
            while current_node._next is not None and exp < current_node._next._exp:
                current_node = current_node._next
            if current_node._next is not None and exp == current_node._next._exp:
                current_node._next._coef += coef
                if current_node._next._coef == 0:
                    current_node._next = current_node._next._next
            else:
                new_node._next = current_node._next
                current_node._next = new_node

    def derivative(self):
        new_poly = Poly()
        current_node = self._head
        while current_node is not None:
            new_coef = current_node._coef * current_node._exp
            new_exp = current_node._exp - 1
            if new_exp >= 0:
                new_poly.insert_node(new_coef, new_exp)
            current_node = current_node._next
        return new_poly
    
    def __mul__(self, other):
        result = Poly()
        current_node1 = self._head
        while current_node1 is not None:
            current_node2 = other._head
            while current_node2 is not None:
                new_coef = current_node1._coef * current_node2._coef
                new_exp = current_node1._exp + current_node2._exp
                result.insert_node(new_coef, new_exp)
                current_node2 = current_node2._next
            current_node1 = current_node1._next
        return result
p1 = Poly()
p1.insert_node(2, 3)
p1.insert_node(4, 2)
p1.insert_node(7, 1)
p1.insert_node(1, 0)
print("p1(x) =", p1)
p2 = Poly()
p2.insert_node(3, 2)
p2.insert_node(5, 1)
p2.insert_node(2, 0)
print("p2(x) =", p2)

print("p1(x) + p2(x) =", p1+p2)

print("d/dx p1(x) =", p1.derivative())

print("p1(x) * p2(x) =", p1 * p2)
```
