```python
class Word:
    def __init__(self, eng, vie):
        self._english = eng
        self._vietnamese = vie

    def __str__(self):
        return self._english + ": " + self._vietnamese

class EVDict:
    def __init__(self):
        self._root = None

    def insert(self, eng, vie):
        word = Word(eng, vie)
        if self._root is None:
            self._root = EVDictNode(word)
        else:
            self._insert_recursive(self._root, word)

    def _insert_recursive(self, node, word):
        if word._english < node._data._english:
            if node._left is None:
                node._left = EVDictNode(word)
            else:
                self._insert_recursive(node._left, word)
        elif word._english > node._data._english:
            if node._right is None:
                node._right = EVDictNode(word)
            else:
                self._insert_recursive(node._right, word)

    def search(self, eng):
        return self._search_recursive(self._root, eng)

    def _search_recursive(self, node, eng):
        if node is None or node._data._english == eng:
            return node
        if eng < node._data._english:
            return self._search_recursive(node._left, eng)
        return self._search_recursive(node._right, eng)

    def delete(self, eng):
        self._root = self._delete_recursive(self._root, eng)

    def _delete_recursive(self, node, eng):
        if node is None:
            return node

        if eng < node._data._english:
            node._left = self._delete_recursive(node._left, eng)
        elif eng > node._data._english:
            node._right = self._delete_recursive(node._right, eng)
        else:
            if node._left is None:
                return node._right
            elif node._right is None:
                return node._left

            min_node = self._find_min_node(node._right)
            node._data = min_node._data
            node._right = self._delete_recursive(node._right, min_node._data._english)

        return node

    def _find_min_node(self, node):
        current = node
        while current._left is not None:
            current = current._left
        return current

    def print_in_order(self):
        self._print_in_order_recursive(self._root)

    def _print_in_order_recursive(self, node):
        if node is not None:
            self._print_in_order_recursive(node._left)
            print(node._data)
            self._print_in_order_recursive(node._right)


class EVDictNode:
    def __init__(self, data):
        self._data = data
        self._left = None
        self._right = None


ev_dict = EVDict()
ev_dict.insert('Hello', 'Xin chào')
ev_dict.insert('Goodbye', 'Tạm biệt')
ev_dict.insert('Apple', 'Quả táo')
ev_dict.insert('Cat', 'Con mèo')
ev_dict.insert('Dog', 'Con chó')

print("Dictionary in alphabetical order:")
ev_dict.print_in_order()

search_word = 'Goodbye'
result = ev_dict.search(search_word)
if result is not None:
    print(f"Meaning of '{search_word}': {result._data._vietnamese}")
else:
    print(f"'{search_word}' not found in the dictionary.")

word_to_delete = 'Apple'
ev_dict.delete(word_to_delete)
print(f"'{word_to_delete}' has been deleted from the dictionary.")

print("Dictionary after deletion:")
ev_dict.print_in_order()

```

```python
class Person:
    def __init__(self, name, year):
        self._name = name
        self._yearOfBirth = year
    
    def __eq__(self, other):
        return self._name == other._name and self._yearOfBirth == other._yearOfBirth
    
    def __str__(self):
        return 'Person[' + self._name + ',' + str(self._yearOfBirth) + ']'

from queue import *
class Tree:
    def __init__(self, data, child=None, sibling=None):
        self._data = data
        self._child = child
        self._sibling = sibling
    
    def dfs(self):
        print(self._data)
        r = self._child
        while r is not None:
            r.dfs()
            r = r._sibling
    
    def bfs(self):
        q = Queue()
        q.put(self)
        while not q.empty():
            r = q.get()
            print(r._data)
            r = r._child
            while r is not None:
                q.put(r)
                r = r._sibling
    
    def count(self):
        d = 1
        r = self._child
        while r is not None:
            d += r.count()
            r = r._sibling
        return d    
    
    def height(self):
        maxh = 0
        r = self._child
        while r is not None:
            h = r.height()
            if h > maxh:
                maxh = h
            r = r._sibling
        return maxh + 1    
    
    def find(self, e):
        if self._data == e:
            return self
        else:
            r = self._child
            while r is not None:
                result = r.find(e)
                if result is not None:
                    return result
                r = r._sibling
            return None    
        
    def findByName(self, name):
        if self._data._name == name:
            return self
        else:
            r = self._child
            while r is not None:
                result = r.findByName(name)
                if result is not None:
                    return result
                r = r._sibling
            return None   
    
    def isParent(self, pname, cname):
        parent = self.findByName(pname)
        if parent is None:
            return False
        else:
            child = parent._child
            while child is not None:
                if child._data._name == cname:
                    return True
                child = child._sibling
            return False
    
    def findLevel(self, name):
        if self._data._name == name:
            return 1
        else:
            r = self._child
            while r is not None:
                result = r.findLevel(name)
                if result > 0:
                    return result + 1
                r = r._sibling
            return 0
    
    def listChildren(self, name):
        person_node = self.findByName(name)
        if person_node is not None:
            child_node = person_node._child
            if child_node is not None:
                print("Những người con của", name, ":")
                while child_node is not None:
                    print(child_node._data._name)
                    child_node = child_node._sibling
            else:
                print(name, "không có người con.")
        else:
            print("Không tìm thấy người", name, "trong cây gia phả.")
    
    def listGeneration(self, generation):
        print("Những người thuộc thế hệ thứ", generation, ":")
        self._listGeneration(self, 1, generation)
    
    def _listGeneration(self, node, level, target_generation):
        if level == target_generation:
            print(node._data._name)
        else:
            child = node._child
            while child is not None:
                self._listGeneration(child, level + 1, target_generation)
                child = child._sibling

    def areSiblings(self, name1, name2):
        person1_node = self.findByName(name1)
        person2_node = self.findByName(name2)
        
        if person1_node is not None and person2_node is not None:
            parent1_node = person1_node._parent
            parent2_node = person2_node._parent
            
            if parent1_node is not None and parent2_node is not None:
                sibling_node = parent1_node._child
                while sibling_node is not None:
                    if sibling_node._data._name == name2:
                        return True
                    sibling_node = sibling_node._sibling
            return False
        else:
            return False

    def addChild(self, parent_name, child_name, child_year):
        parent_node = self.findByName(parent_name)
        if parent_node is not None:
            child_node = Tree(Person(child_name, child_year))
            if parent_node._child is None:
                parent_node._child = child_node
            else:
                sibling_node = parent_node._child
                while sibling_node._sibling is not None:
                    sibling_node = sibling_node._sibling
                sibling_node._sibling = child_node
            print("Đã thêm người", child_name, "là con của", parent_name)
        else:
            print("Không tìm thấy người", parent_name, "trong cây gia phả.")





# Create the tree structure
n1 = Tree(Person('Nguyen H', 1970))
n2 = Tree(Person('Nguyen G', 1965), None, n1)
n3 = Tree(Person('Nguyen C', 1935), n2, None)
root = Tree(Person('Nguyen A', 1900), n3, None)

# Test the methods

# Count the number of people in the family tree
num_people = root.count()
print("Number of people in the family tree:", num_people)

# Calculate the number of generations in the family tree
num_generations = root.height()
print("Number of generations in the family tree:", num_generations)

# Find a person by name in the family tree
person_name = "Nguyen G"
person_node = root.findByName(person_name)
if person_node is not None:
    print("Found person:", person_node._data)
else:
    print("Person", person_name, "not found in the family tree.")

# Check if a person named 'Nguyen H' is a child of a person named 'Nguyen C'
parent_name = "Nguyen C"
child_name = "Nguyen H"
is_parent = root.isParent(parent_name, child_name)
if is_parent:
    print(child_name, "is a child of", parent_name)
else:
    print(child_name, "is not a child of", parent_name)

# Calculate the generation of a person by name in the family tree
person_name = "Nguyen G"
generation = root.findLevel(person_name)
if generation > 0:
    print(person_name, "is in generation", generation)
else:
    print("Person", person_name, "not found in the family tree.")

# List all the children of a person named 'Nguyen C'
parent_name = "Nguyen C"
root.listChildren(parent_name)

# List all the people in generation 2
generation_number = 2
root.listGeneration(generation_number)

# Check if two people named 'Nguyen H' and 'Nguyen I' are siblings
person1_name = "Nguyen H"
person2_name = "Nguyen I"
are_siblings = root.areSiblings(person1_name, person2_name)
if are_siblings:
    print(person1_name, "and", person2_name, "are siblings.")
else:
    print(person1_name, "and", person2_name, "are not siblings.")

# Add a child named 'Nguyen Q' to a person named 'Nguyen C'
parent_name = "Nguyen C"
child_name = "Nguyen Q"
child_year = 2000  # Replace with the actual birth year of Nguyen Q
root.addChild(parent_name, child_name, child_year)
```
