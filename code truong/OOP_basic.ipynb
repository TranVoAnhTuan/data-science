```python
class Triangle:
    """Khai bao thuoc tinh"""
    _c1 = 0.0
    _c2 = 0.0
    _c3 = 0.0
    
    
    def __init__(self, a, b, c):
        """Khai bao phuong thuc"""
        if (a + b > c) or (a + c > b) or (b + c > a):
            self._c1 = a
            self._c2 = b
            self._c3 = c
            
    
    def perimeter(self):
        """Phuong thuc tinh chu vi"""
        return self._c1 + self._c2 + self._c3
    
    
    def area(self):
        """Phuong thuc tinh dien tich"""
        half_perimeter = self.perimeter() / 2   #tinh nua chu vi
        area = (half_perimeter * (half_perimeter - self._c1) * (half_perimeter - self._c2) * (half_perimeter - self._c3)) ** 0.5    #tinh dien tich
        return area
        
    
    def __str__(self):
        """Phuong thuc chuyen thanh xau"""
        return f"Triangle: _c1 = {self._c1}, _c2 = {self._c2}, _c3 = {self._c3}"
    
    
    def classify_triangle(self):
        """Phuong thuc phan loai tam giac"""
        sides = sorted([self._c1, self._c2, self._c3])  # Lay 3 cot duoc sap xep tang dan
        if sides[2] ** 2 == sides[0] ** 2 + sides[1] ** 2:  # Kiem tra dieu kien de phan loai
            print("Triangle is a right triangle")
        elif sides[0] == sides[1] == sides[2]:
            print("Triangle is equilateral")
        elif sides[0] == sides[1] or sides[1] == sides[2] or sides[0] == sides[2]:
            print("Triangle is isosceles")
        else:
            print("Triangle is scalene")

# Tao danh sach 5 tam giac
triangles = [
    Triangle(3, 4, 5),
    Triangle(5, 12, 13),
    Triangle(6, 8, 10),
    Triangle(5, 5, 5),
    Triangle(6, 6, 8)
]

# In ra thong tin
total_area = 0
for triangle in triangles:
    print(triangle)
    print(f"Chu vi: {triangle.perimeter()}")
    print(f"Dien tich: {triangle.area()}")
    triangle.classify_triangle()
    total_area += triangle.area()
    print("------------------------")

print(f"Tong dien tich cua 5 tam giac la: {total_area}")
maxArea = float('-inf')
maxPerimeter = float('-inf')
for triangle in triangles:
    if triangle.perimeter() > maxPerimeter:
        maxPerimeter = triangle.perimeter()
    if triangle.area() > maxArea:
        maxArea = triangle.area()
print(f'max perimeter: {maxPerimeter}; max area: {maxArea}')    

```

```python
class Vector:
    def __init__(self, lst):
        self._values = list(lst)
        
    def __add__(self, another):
        assert len(self) == len(another), "Error in adding. Length of vectors must be same."
        return Vector([a + b for a, b in zip(self, another)])

    def __sub__(self, another):
        assert len(self) == len(another), "Error in subtracting. Length of vectors must be same."
        return Vector([a - b for a, b in zip(self, another)])

    def __neg__(self):
        return Vector([-a for a in self])

    def __mul__(self, k):
        if isinstance(k, int):
            return Vector([k * a for a in self])
        else:
            assert len(self) == len(k), "Error in multiplying. Length of vectors must be same."
            return sum([a * b for a, b in zip(self, k)])
    
    def __rmul__(self, k):
        return self * k
    
    def __iter__(self):
        return self._values.__iter__()

    def __len__(self):
        return len(self._values)
    
```
