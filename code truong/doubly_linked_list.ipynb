```python

class Node:
    def __init__(self, data):
        self._data = data
        self._next = None

class CircularLinkedList:
    def __init__(self):
        self._head = None

    def is_empty(self):
        return self._head is None

    def add_first(self, e):
        new_node = Node(e)
        if self._head == None:
            new_node._next = new_node
            self._head = new_node
        else:
            last = self._head
            while last._next != self._head: 
                last = last._next
        last._next = new_node
        new_node._next = self._head
        self._head = new_node

    def add_last(self, e):
        new_node = Node(e)
        if self._head is None:
            self._head = new_node
            new_node._next = self._head
        else:
            last = self._head
            while last._next != self._head:
                last = last._next
            last._next = new_node
            new_node._next = self._head

    def remove_first(self):
        if self._head == None: 
            return
        if self._head._next == self._head:	
            self._head = None
        else:
            last = self._head
            while last._next != self._head: 
                last = last._next
            last._next = self._head._next
            self._head = last._next

    def remove_last(self):
        if self._head == None: 
            return
        if self._head._next == self._head:
            self._head = None
        else:
            plast = self._head
            while plast._next._next != self._head: 
                plast = plast._next
            plast._next = self._head
        
    def display(self):
        if self._head is None:
            print("List is empty")
        else:
            current = self._head
            while current._next != self._head:
                print(current._data, end= '\t')
                current = current._next
            print(current._data, end= '\n')

    def count(self):
        count = 0
        current = self._head
        while current:
            count += 1
            current = current._next
            if current == self._head:
                break
        return count
    
    def to_array(self):
        arr = []
        current = self._head
        while current:
            arr.append(current._data)
            current = current._next
            if current == self._head:
                break
        return arr
    
    def makeLinkedList(self, a):
        for i in a:
            self.add_last(i)

    def mergeList(self, other):
        if self._head is None and other._head is None:
            return None
        elif self._head is None:
            return other
        elif other._head is None:
            return self
        
        newLinkedList = CircularLinkedList()
        newLinkedList.add_last(self._head._data)
        newLinkedList.add_last(other._head._data)
        currentCLL1 = self._head._next
        currentCLL2 = other._head._next
        while currentCLL1 != self._head and currentCLL2 != other._head:
            newLinkedList.add_last(currentCLL1._data)
            newLinkedList.add_last(currentCLL2._data)
            currentCLL1 = currentCLL1._next
            currentCLL2 = currentCLL2._next
        while currentCLL1 != self._head:
            newLinkedList.add_last(currentCLL1)
            currentCLL1 = currentCLL1._next
        while currentCLL2 != other._head:
            newLinkedList.add_last(currentCLL2)
            currentCLL2 = currentCLL2._next
        return newLinkedList


CLL1 = CircularLinkedList()
CLL1.add_last(1)
CLL1.add_last(2)
CLL1.add_last(3)
CLL1.display()

CLL1.remove_first()
CLL1.display() 

CLL1.remove_last()
CLL1.display()

print(CLL1.count())

arr = CLL1.to_array()
print(arr)

test1 = CircularLinkedList()
test1.makeLinkedList([1,2,3])
test1.display()
test2 = CircularLinkedList()
test2.makeLinkedList([4, 5, 6])
test2.display()

lst2 = test1.mergeList(test2)
lst2.display()
```

```python
class DNode:
    def __init__(self, e, prev = None, next = None):
        self._element = e
        self._prev = prev
        self._next = next
       
class DoublyLinkedList:
    def __init__(self):
        self._head = None
        self._tail = None
        self._size = 0
    
    def add_first(self, e):
        new_node = DNode(e, None, self._head)
        if self._head is None:
            self._tail = new_node
        else:
            self._head._prev = new_node
        self._head = new_node
        self._size += 1

    def remove_first(self):
        if self._head is None:
            return
        if self._head == self._tail:
            self._head = None
            self._tail = None
        else:
            self._head = self._head._next
            self._head._prev = None
        self._size -= 1    

    def remove(self, e):
        if self._head is None:
            return
        if self._head._element == e:
            self.remove_first()
        else:
            current = self._head
            while current is not None and current._element != e:
                current = current._next
            if current is not None:
                pcurrent = current._prev
                ncurrent = current._next
                pcurrent._next = ncurrent
                if ncurrent is not None:
                    ncurrent._prev = pcurrent
            if current == self._tail:
                self._tail = pcurrent
            self._size -= 1
                           
    def display(self):
        current = self._head
        while current is not None:
            print(current._element,end= '\t')
            current = current._next
        print('\n')

    def is_symmetric(self):
        if self._head is None:
            return True
        left = self._head
        right = self._tail
        while left != right and left._prev != right:
            if left._element != right._element:
                return False
            left = left._next
            right = right._prev
        return True

    
    def insert_sort(self):
        if self._head is None or self._head._next is None:
            return
        new_list = DoublyLinkedList()
        new_list.add_first(self._head._element)
        current = self._head._next
        while current is not None:
            insert_after = new_list._tail
            while insert_after is not None and insert_after._element > current._element:
                insert_after = insert_after._prev
            if insert_after is None:
                new_list.add_first(current._element)
            else:
                new_node = DNode(current._element, insert_after, insert_after._next)
                insert_after._next = new_node
                if new_node._next is not None:
                    new_node._next._prev = new_node
                else:
                    new_list._tail = new_node
            current = current._next
        self._head = new_list._head
        self._tail = new_list._tail

linked_list = DoublyLinkedList()
linked_list.add_first(3)
linked_list.add_first(4)
linked_list.add_first(1)
linked_list.add_first(5)
linked_list.add_first(2)
linked_list.display()
print(linked_list.is_symmetric())
linked_list.insert_sort()
linked_list.display()
```
