# 2

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
ten = ['Trung','Cnh','Bo','Quc','Tng']
vangLT = [3,5,1,0,2]
vangTH = [2,1,0,0,1]
khongNB = [0,3,0,2,1]
duLieu = zip(ten,vangLT,vangTH,khongNB)
cot=['Tn sinh vin', 'Vng l thuyt', 'Vng thc hnh','Khng np bi']
df = pd.DataFrame(data = duLieu, columns=cot)
df['Tng li'] = df['Vng l thuyt'] + df['Vng thc hnh'] + df['Khng np bi']
df
plt.pie(df['Tng li'], labels=df['Tn sinh vin'], explode=(0,0.15,0,0,0), startangle=200, autopct='%1.1f%%')
plt.legend()
plt.axis('equal')
plt.show()
```
