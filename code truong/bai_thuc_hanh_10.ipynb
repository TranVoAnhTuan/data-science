```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
ten = ['Trung','Cảnh','Bảo','Quốc','Tùng']
vangLT = [3,5,1,0,2]
vangTH = [2,1,0,0,1]
khongNB = [0,3,0,2,1]
duLieu = zip(ten,vangLT,vangTH,khongNB)
cot=['Tên sinh viên', 'Vắng lý thuyết', 'Vắng thực hành','Không nộp bài']
df = pd.DataFrame(data = duLieu, columns=cot)
df['Tổng lỗi'] = df['Vắng lý thuyết'] + df['Vắng thực hành'] + df['Không nộp bài']
df
plt.pie(df['Tổng lỗi'], labels=df['Tên sinh viên'], explode=(0,0.15,0,0,0), startangle=200, autopct='%1.1f%%')
plt.legend()
plt.axis('equal')
plt.show()
```
